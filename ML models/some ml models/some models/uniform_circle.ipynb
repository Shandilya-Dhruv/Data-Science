```python
"""uniform_circle.ipynb generates uniformly distributed random points inside of a circle"""
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def generate_random(n_points = 100, centre = (0,0), radius = 10):

    l = []

    for i in range(n_points):

        theta_rand = random.uniform(0,2*np.pi)
        r_rand = radius*np.sqrt(random.uniform(0,1))

        p = [0,0]

        p[0] = centre[0] + r_rand*np.cos(theta_rand)
        p[1] = centre[1] + r_rand*np.sin(theta_rand)

        l.append((p[0],p[1]))
    
    return l
```

```python
def plot(l):
    
    for i in l:
        plt.scatter(i[0],i[1])

    plt.show()
```

```python
l = generate_random(1000,(0,0),10)
```

```python
plot(l)
```
